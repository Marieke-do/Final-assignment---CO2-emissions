```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

%load_ext google.colab.data_table
```

```python
# Read CO2 change in % csv
df_co2_perc = pd.read_csv('https://raw.githubusercontent.com/Marieke-do/Final-assignment---CO2-emissions/main/change-co2-annual-pct.csv')
df_co2_perc.drop(columns = 'Code', inplace=True)
df_co2_perc.rename(columns = {'Annual CO2 emissions growth (%)': 'CO2 change %'}, inplace=True)
df_co2_perc

# Read CO2 absolute change csv
df_co2_abs = pd.read_csv('https://raw.githubusercontent.com/Marieke-do/Final-assignment---CO2-emissions/main/absolute-change-co2.csv')
df_co2_abs.drop(columns = 'Code', inplace = True)
df_co2_abs.rename(columns = {'Annual CO2 emissions growth (abs)': 'CO2 change abs'}, inplace =True)
df_co2_abs

# Read total annual CO2 emissions per country
df_co2_total = pd.read_csv('https://raw.githubusercontent.com/Marieke-do/Final-assignment---CO2-emissions/main/annual-co2-emissions-per-country.csv')
df_co2_total.drop(columns = 'Code', inplace =True)

# Read CO2 emissions per capita csv
df_co2 = pd.read_csv('https://raw.githubusercontent.com/Marieke-do/Final-assignment---CO2-emissions/main/co2-emissions-vs-gdp.csv')
df_co2.drop(columns = ['Code', 'GDP per capita','145446-annotations', 'Continent'], inplace=True)
df_co2.rename(columns = {'Annual CO2 emissions (per capita)':'Annual CO2 per capita', 'Total population (Gapminder, HYDE & UN)': 'Total population'}, inplace=True)
df_co2

# Merge DataFrames
df_co2_perc_abs = pd.merge(df_co2_abs, df_co2_perc, on=['Entity', 'Year'])
df_co2_total_per_cap = pd.merge (df_co2_total, df_co2, on=['Entity', 'Year'])
df = pd.merge(df_co2_total_per_cap, df_co2_perc_abs, on=['Entity', 'Year'])

# Change order of columns

cols = df.columns.to_list()
new_cols = cols[0:5] + cols[6:] + cols[5:6]
df_def = df.loc[:, new_cols]

df_def


```

In 2016 the Paris Agreement was signed by 196 parties. "*The Paris Agreement's long-term temperature goal is to keep the rise in mean global temperature to well below 2 °C (3.6 °F) above pre-industrial levels, and preferably limit the increase to 1.5 °C (2.7 °F), recognising that this would substantially reduce the impacts of climate change. Emissions should be reduced as soon as possible and reach net-zero by the middle of the 21st century.*"(https://en.wikipedia.org/wiki/Paris_Agreement)

---
To answer question 2 (Which countries are making the biggest strides in decreasing CO2 output?), the period after the Agreement has been signed has been chosen to work with: 2016-2019 (data from 2020 and 2021 are not availble yet).

```python
# Filter data by years 2016-2019 and drop some more columns

years = [2016, 2019]
df_by_year = df_def.loc[df["Year"].isin(years)]
df_by_year.drop(columns = ['Total population', 'CO2 change abs', 'CO2 change %'], inplace = True)
df_by_year
```

```python
df_by_year['2016-2019 abs change'] = df_by_year.loc[:,'Annual CO2 emissions'].diff()
df_by_year['2016-2019 % change'] = df_by_year.loc[:,'Annual CO2 emissions'].pct_change()
df_by_year['2016-2019 abs change per capita'] = df_by_year.loc[:,'Annual CO2 per capita'].diff()
df_by_year['2016-2019 % change per capita'] = df_by_year.loc[:,'Annual CO2 per capita'].pct_change()
df_2019= df_by_year.loc[df_by_year['Year'] == 2019]
df_2019
```

```python
# Top 10 biggest decreasers in absolute CO2

sorted_total = df_2019.sort_values('2016-2019 abs change')
 
sorted_total = sorted_total[~sorted_total['Entity'].str.contains('EU-28|EU-27|Europe|South America|North America')]

sorted_total.iloc[0:10]
```
